```python
# Hello World !
# Perceptron !!

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris  # 3 انواع من الورود 
from sklearn.linear_model import Perceptron
```

```python
iris = load_iris(return_X_y=True, as_frame=True)
# return_X_y=True:
# بدون هذه الإضافة: المكتبة ستعطيك "قاموس" (Dictionary) يحتوي على وصف للبيانات، وأسماء الأعمدة، والبيانات نفسها.
# مع هذه الإضافة: أنت تقول لها "أعطيني فقط البيانات (X) والنتائج (y) بشكل منفصل ومباشر". وهذا يسهل عليك تمريرها مباشرة للموديل.

# as_frame=True:
# بدون هذه الإضافة: ستكون البيانات على شكل مصفوفة NumPy (أرقام فقط بدون أسماء أعمدة).
# مع هذه الإضافة: يتم تحميل البيانات كـ Pandas DataFrame. هذا رائع لأنه يسمح لك برؤية أسماء الأعمدة (مثل sepal length (cm)) والتعامل معها وكأنها جدول Excel داخل VS Code.
```

```python
print(iris)
data = iris[0]
target = iris[1]
```

```python
df = pd.concat([data, target], axis=1)
df.head() 
```

```python
df = df.iloc[:100, [0, 2, -1]]
print(df)
```

```python
x = df.iloc[:, [0, 1]].values # sepal length | petal length
y = df.iloc[:, -1].values  # target
```

```python
# الرسم 
plt.scatter(x[:50, 0], x[:50, 1], color='red', marker='o', label='Setosa')
plt.scatter(x[50:100, 0], x[50:100, 1], color='purple', marker='s', label='Versicolor')
plt.legend(loc='upper left')
plt.xlabel("Sepal lengh (cm)")
plt.ylabel("Petal lengh (cm)")
plt.show()
```

```python
model = Perceptron()
model.fit(x, y)
# !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
```
