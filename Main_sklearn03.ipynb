```python
import seaborn as sb

import pandas as pd

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
titanic = pd.read_csv(url)


sb.set_theme(style="darkgrid", palette="bright")

sb.load_dataset('titanic')
sb.load_dataset('iris')
tips = sb.load_dataset('tips')
sb.load_dataset('planets')
```

```python
sb.scatterplot(x='tip', y='total_bill', data=tips, hue ="day", size="size", palette="YlGnBu")
```

```python
sb.histplot(tips['tip'], kde=True, bins=15) # kde هو ليدير الخط 
```

```python
sb.displot(tips['tip'], kde=True)
```

```python
sb.barplot(x="sex", y="tip", data =tips, palette="YlGnBu")
```

```python
sb.boxplot(x='day', y='tip', data=tips , hue="sex", palette="YlGnBu")
```

```python
sb.stripplot(x='day', y='tip', data=tips, hue="sex", palette="YlGnBu")
```

```python
sb.stripplot(x='day', y='tip', data=tips, hue="sex", palette="YlGnBu", dodge=True)
# dodge for separate the histogram
```

```python
sb.jointplot(x='tip', y='total_bill', data=tips)
```

```python
sb.jointplot(x='tip', y='total_bill', data=tips, kind='reg')
```

```python
sb.jointplot(x='tip', y='total_bill', data=tips, kind='kde')
```

```python
sb.jointplot(x='tip', y='total_bill', data=tips, kind='kde', shade=True, cmap='YlGnBu')
```

```python
sb.jointplot(x='tip', y='total_bill', data=tips, kind='hex', cmap='YlGnBu')
```
